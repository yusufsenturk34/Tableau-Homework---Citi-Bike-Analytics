```python
import pandas as pd
import glob
```

```python
interesting_files = glob.glob("*.csv") 
df = pd.concat((pd.read_csv(f, header = 0) for f in interesting_files))
df
```

```python
df['gender'].replace(2, 'Female',inplace=True)
df['gender'].replace(1, 'Male',inplace=True)
df['gender'].replace(0, 'Unknown', inplace=True)
```

```python
df.to_csv("2019-citibike-trip-data.csv")
```
